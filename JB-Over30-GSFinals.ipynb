```python
import pandas as pd 
```

### grabbed this from https://github.com/JeffSackmann/tennis_atp

```python


matches = pd.read_csv('https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_2015.csv')
```

```python
matches.shape
```

```python
#making a list of urls 
urls = []
for year in range(1969, 2017):
    url_template = 'https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_'+str(year)+'.csv'
    urls.append(url_template)
```

```python
##reading those urls and combining them into a dataframe to manipulate 

#I started by grabbing 1968 and then used a for loop to grab the other years 

#this took about 4 minutes to run. 

df_all = pd.read_csv('https://raw.githubusercontent.com/JeffSackmann/tennis_atp/master/atp_matches_1968.csv')
for year in urls:
    df = pd.read_csv(year)
    df_all = pd.concat([df_all,df])
```

```python
##checking out the data - these are wimbledon finals 

df_all[(df_all.tourney_name == 'Wimbledon') & (df_all['round'] =='F' ) ]
```

```python
##grabbing only grand-slam tournaments 

all_GS = df_all[(df_all.tourney_name == 'US Open')|(df_all.tourney_name == 'Wimbledon')|(df_all.tourney_name == 'Australian Open')|(df_all.tourney_name == 'Roland Garros')]
```

```python
##grabbing all finals 

GS_F = all_GS[all_GS['round']=='F']
```

```python
##grabbing GS finals where both players are above 30 

GS_F30 = GS_F[(GS_F.winner_age >= 30.0) & (GS_F.loser_age >= 30.0)]
```

# Results below 

```python
##yep looks like the 1969 Aussie and French both won by Rod Laver, 1972 Aussie, and 2002 U.S. Open

GS_F30
```

```python
print('30+ Winners:', GS_F30.winner_name.value_counts())
print('30+ Losers:', GS_F30.loser_name.value_counts())
```
